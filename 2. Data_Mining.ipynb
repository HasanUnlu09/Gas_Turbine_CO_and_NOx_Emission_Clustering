# Data_Mining

---

Data is obtained from [UCI ML Repository - Gas Turbine CO and NOx Emission Data Set](https://archive.ics.uci.edu/dataset/551/gas+turbine+co+and+nox+emission+data+set) with .csv extension.\
\
Citation:\
Gas Turbine CO and NOx Emission Data Set. (2019). UCI Machine Learning Repository. \
https://doi.org/10.24432/C5WC95.

---

```python
import pandas as pd
```

```python
# reading all datasets
data_2011 = pd.read_csv('gt_2011.csv')
data_2012 = pd.read_csv('gt_2012.csv')
data_2013 = pd.read_csv('gt_2013.csv')
data_2014 = pd.read_csv('gt_2014.csv')
data_2015 = pd.read_csv('gt_2015.csv')
```

```python
# all the datasets are converted into one big dataset
data = data_2011.append([data_2012, data_2013, data_2014, data_2015], ignore_index=True)
```

```python
data
```

```python
# the merged dataset is stored in the below file
data.to_csv('CO_NOx_Emissions.csv', index=None)
```
