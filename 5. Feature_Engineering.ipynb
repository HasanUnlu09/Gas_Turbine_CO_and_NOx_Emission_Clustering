## 3. Feature Scaling

```python
# Necessary library
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# the dataset before scaling
sns.displot(cleaned_data, aspect=3, kde=True)
```

---

```python
# Standardization
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
standardized = cleaned_data.copy(deep=True)
standardized[:] = standard_scaler.fit_transform(standardized[:])
standardized = pd.DataFrame(standardized)
sns.displot(standardized, aspect=3, kde=True)
```

```python
[print(standardized.loc[:,j].min(), standardized.loc[:,j].max()) for i,j in enumerate(standardized.columns)]
```

```python
standardized.shape
```

---

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
normalized = cleaned_data.copy(deep=True)
normalized[:] = min_max_scaler.fit_transform(normalized[:])
normalized = pd.DataFrame(normalized)
sns.displot(normalized, aspect=3, kde=True)
```

```python
[print(normalized.loc[:,j].min(), normalized.loc[:,j].max()) for i,j in enumerate(normalized.columns)]
```

```python
normalized.shape
```

---

```python
# Maximum Absolute Scaling
from sklearn.preprocessing import MaxAbsScaler
max_abs_scaler = MaxAbsScaler()
max_abs_scaled = cleaned_data.copy(deep=True)
max_abs_scaled[:] = max_abs_scaler.fit_transform(max_abs_scaled[:])
max_abs_scaled = pd.DataFrame(max_abs_scaled)
sns.displot(max_abs_scaled, aspect=3, kde=True)
```

```python
[print(max_abs_scaled.loc[:,j].min(), max_abs_scaled.loc[:,j].max()) for i,j in enumerate(max_abs_scaled.columns)]
```

```python
max_abs_scaled.shape
```

---

```python
# Robust Scaling
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
robust_scaled = cleaned_data.copy(deep=True)
robust_scaled[:] = robust_scaler.fit_transform(robust_scaled[:])
robust_scaled = pd.DataFrame(robust_scaled)
sns.displot(robust_scaled, aspect=3, kde=True)
```

```python
[print(robust_scaled.loc[:,j].min(), robust_scaled.loc[:,j].max()) for i,j in enumerate(robust_scaled.columns)]
```

```python
robust_scaled.shape
```

---

```python
# Quantile Transformer Scaling
from sklearn.preprocessing import QuantileTransformer
quantile_transformer = QuantileTransformer()
quantile_transformed = cleaned_data.copy(deep=True)
quantile_transformed[:] = quantile_transformer.fit_transform(quantile_transformed[:])
quantile_transformed = pd.DataFrame(quantile_transformed)
sns.displot(quantile_transformed, aspect=3, kde=True)
```

```python
[print(quantile_transformed.loc[:,j].min(), quantile_transformed.loc[:,j].max()) for i,j in enumerate(quantile_transformed.columns)]
```

```python
quantile_transformed.shape
```

---

```python
# Power Transformer Scaling (with Yeo-Johnson Transform)
from sklearn.preprocessing import power_transform
powered = cleaned_data.copy(deep=True)
powered[:] = power_transform(powered[:], method='yeo-johnson')
powered = pd.DataFrame(powered)
sns.displot(powered, aspect=3, kde=True)
```

```python
[print(powered.loc[:,j].min(), powered.loc[:,j].max()) for i,j in enumerate(powered.columns)]
```

```python
powered.shape
```

## 4. Storing The Data

The data that is applied feature engineering method is stored in this section.

```python
# Storing the transformed data

standardized.to_csv('Transformed_Data_Standardized.csv', index=False)
normalized.to_csv('Transformed_Data_Normalized.csv', index=False)
robust_scaled.to_csv('Transformed_Data_Robust.csv', index=False)
powered.to_csv('Transformed_Data_Powered.csv', index=False)
```
